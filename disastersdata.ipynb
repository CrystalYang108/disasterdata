Creating new dataframe to have only those columns,
Creating new dataframe where only columns with specific indcident type and specific year

```python
df = pd.read_csv('https://raw.githubusercontent.com/CrystalYang108/disasterdata/main/DisasterDeclarationsSummaries.csv')
df2=df[['fyDeclared','state','fipsStateCode','incidentType','disasterNumber']]
df2=df2[(df2['incidentType']=='Flood') & (df['fyDeclared']==2022)]
df2.head()
```

Drops duplicate values such as repeated states with the same incident type and disaster number values

```python
df_dup = df2.drop_duplicates(subset=["state"])
df_dup.head(10)
```

Visualization of new dataframe

```python
import plotly.express as py
fig = py.choropleth(df_dup,
                    locations='state', 
                    locationmode="USA-states", 
                    scope="usa",
                    color='disasterNumber',
                    color_continuous_scale="blues", 
                    title = 'Number of Floods in the United States 2022'
                    
                    )
fig.show()
```
